```python
import plotly.graph_objects as go
import pandas as pd
import plotly.io as pio
import os
```

### Data

```python
df_R1_DE = pd.DataFrame({
    "Soluplus": [5.90,50,29.3,50,36.50,50,5,29.75,29.75,5,29.30,5,21.43,13.78,50],
    "TPGS": [27.52,0.05,50,50,0.05,20.78,0.05,27.52,1.05,50,27.04,0.05,50,50,18.53],
    "DTX": [2.65,1,1.60,2.13,4,2.22,4,1.06,2.65,1,3.25,1,4,2.65,4]
})

df_R2_DE = pd.DataFrame({
    "Soluplus": [50,50,49.99,50,50,50,49.73,50,49.658,49.996,49.176,50,50,49.383,49.167],
    "TPGS": [0.05, 0.05, 0.05, 0.254, 0.334, 0.05, 0.05, 0.05, 0.05, 0.05, 0.335, 0.969, 0.05, 0.402, 0.05],
  "DTX": [2.25, 2.251, 2.275, 2.261, 2.265, 2.307, 2.26, 2.325, 2.263, 2.345, 2.275, 2.299, 2.432, 2.29, 2.281]
})

df_R3_DE = pd.DataFrame({
    "Soluplus": [49.062, 49.15, 49.135, 7.862, 49.231, 49.098, 49.066, 8.299, 49.046, 49.303, 7.866, 49.154, 7.807, 47.518, 48.594],
    "TPGS": [1.403, 0.162, 0.723, 0.856, 1.763, 1.503, 20.428, 0.852, 2.095, 0.526, 1.132, 0.301, 2.647, 22.249, 18.071],
    "DTX": [3.346, 3.126, 3.173, 1.968, 3.705, 3.425, 1.136, 2.631, 3.37, 3.95, 2.046, 3.164, 1.999, 1.033, 1.058]
})


df_R1_DE["Round"] = "Round 1"
df_R2_DE["Round"] = "Round 2"
df_R3_DE["Round"] = "Round 3"

# Merge DataFrame
df_all = pd.concat([df_R1_DE, df_R2_DE, df_R3_DE], ignore_index=True)
```

### Plot

```python
# ------------------------
# Colors and symbols
# ------------------------
colors = {
    "Round 1": "#4C78A8",
    "Round 2": "#F58518",
    "Round 3": "#54A24B"
}

symbols = {
    "Round 1": "circle",
    "Round 2": "square",
    "Round 3": "diamond"
}

# ------------------------
# Plot
# ------------------------
fig = go.Figure()

for round_name in df_all["Round"].unique():
    
    df = df_all[df_all["Round"] == round_name]

    fig.add_trace(go.Scatter3d(
        x=df["Soluplus"],
        y=df["TPGS"],
        z=df["DTX"],
        mode='markers',
        marker=dict(
            size=6,
            color=colors[round_name],
            symbol=symbols[round_name],
            line=dict(color='black', width=0.8),
            opacity=0.9
        ),
        name=round_name
    ))

fig.update_layout(
    title="Design space points in different rounds of Design Expert",
    scene=dict(
        xaxis=dict(title="Soluplus (mg/mL)", range=[0, 50], showgrid=True,
                   gridcolor='rgba(120,120,120,0.35)'),
        yaxis=dict(title="TPGS (mg/mL)", range=[0, 50], showgrid=True,
                   gridcolor='rgba(120,120,120,0.35)'),
        zaxis=dict(title="DTX (mg/mL)", range=[0, 5], showgrid=True,
                   gridcolor='rgba(120,120,120,0.35)'),
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.2))
    ),
    #template="plotly_white",
    width=900,
    height=700
)

pio.renderers.default = "vscode"

fig.show()
```

### Plotly graph depending on each predicted variable 

#### Round 1

```python
#Route to the file

BASE_DIR = os.getcwd()

file_path = os.path.join(
    BASE_DIR,
    "Pred_obs_R1.csv",
)

#Read the CSV file
df_R1 = pd.read_csv(file_path, sep=";")
df_R1 = df_R1.dropna(how="all") # Remove rows where all values are NaN
```

```python
variables_pred = ["Size_pred", "PdI_pred", "ZP_pred", "EE_pred", "DL_pred"]
```

```python
for var in variables_pred:

    fig = go.Figure()

    fig.add_trace(go.Scatter3d(
        x=df_R1["Soluplus_mg_ml"],
        y=df_R1["TPGS_mg_ml"],
        z=df_R1["DTX_mg_ml"],
        mode='markers',
        marker=dict(
            size=6,
            color=df_R1[var], 
            colorscale='Viridis',
            colorbar=dict(title=var),
            line=dict(color='black', width=0.5),
            opacity=0.9
        ),
        name=var
    ))

    fig.update_layout(
        title=f"Design space coloured by {var}",
        scene=dict(
            xaxis=dict(title="Soluplus (mg/mL)"),
            yaxis=dict(title="TPGS (mg/mL)"),
            zaxis=dict(title="DTX (mg/mL)")
        ),
        width=900,
        height=700
    )

    fig.show()
```

```python
variables_obs = ["Size_obs", "PdI_obs", "ZP_obs", "EE_obs", "DL_obs"]
```

```python
for var in variables_obs:

    fig = go.Figure()

    fig.add_trace(go.Scatter3d(
        x=df_R1["Soluplus_mg_ml"],
        y=df_R1["TPGS_mg_ml"],
        z=df_R1["DTX_mg_ml"],
        mode='markers',
        marker=dict(
            size=6,
            color=df_R1[var], 
            colorscale='Viridis',
            colorbar=dict(title=var),
            line=dict(color='black', width=0.5),
            opacity=0.9
        ),
        name=var
    ))

    fig.update_layout(
        title=f"Design space coloured by {var}",
        scene=dict(
            xaxis=dict(title="Soluplus (mg/mL)"),
            yaxis=dict(title="TPGS (mg/mL)"),
            zaxis=dict(title="DTX (mg/mL)")
        ),
        width=900,
        height=700
    )

    fig.show()
```
